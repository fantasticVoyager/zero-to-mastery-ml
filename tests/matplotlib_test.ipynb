#### Matplotlib Example Workflow

```python
# Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#prepare data
x1 = [1, 2, 3, 4]
y1 = [11, 22, 33, 44]

#setup plot
fig1, ax1 = plt.subplots(figsize=(10, 10))

#plot data
ax1.plot(x1, y1)

#customize plot
ax1.set(title="Sample Plot",
        xlabel="x label",
        ylabel="y label")

#save and show plot
fig1.savefig("./simple-plot.png")
```

### Multiple plots in a figure, 2 options

```python
#Option 1 for subplot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
#Option 2 for Subplots
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
#Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

```python
heart_disease = pd.read_csv("../data/heart-disease.csv")
heart_disease.head()
```

```python
#histogram to see distribution of the age column
heart_disease["age"].plot.hist()
```

```python
heart_disease["age"].mean()
```

```python
heart_disease["age"].std()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#this is the pyplot method where we plot directly from the DataFrame
#c is for color so we are coloring the dot based on the target column
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target")
```

```python
#this is the OO/pyplot way to do the same thing, we have more conntrol
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
#the ax parameter is passed the axis that we created with subplots, already have a bit more control
ax.set_xlim([45, 100]) #sets the min and max for the x axis
```

```python
#OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(), title="Target");

#Add a horizontal line at the mean of cholesterol levels
ax.axhline(over_50["chol"].mean(),
           linestyle="dashed")
```

## Using the OOP method to create subplots

```python
fig, (ax0, ax1) = plt.subplots(figsize=(10, 10),
                               nrows=2,
                               ncols=1,
                               sharex=True)

#Add data to axes
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

#Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

ax0.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(over_50["chol"].mean(),
            linestyle="--")

scatter2 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"])

ax1.set(title="Heart Disease and Thalach",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.legend(*scatter2.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--")

#Add a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16)
```

```python
#Putting it all together for a nice looking and useful pl9t
plt.style.use('seaborn-whitegrid')

fig, (ax0, ax1) = plt.subplots(figsize=(10, 10),
                               nrows=2,
                               ncols=1,
                               sharex=True)

#Add data to axes
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")

#Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")

ax0.set_xlim([50, 80])

ax0.legend(*scatter.legend_elements(), title="Target")

ax0.axhline(over_50["chol"].mean(),
            linestyle="--")

scatter2 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"],
                       cmap="winter")

ax1.set(title="Heart Disease and Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

ax1.legend(*scatter2.legend_elements(), title="Target")
ax1.axhline(over_50["thalach"].mean(),
            linestyle="--")

#Add a title to the entire figure
fig.suptitle("Heart Disease Analysis", fontsize=16)
```
